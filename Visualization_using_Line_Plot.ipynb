```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create the data
Day=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
New_Births=[12.15,11,9,1,9,21, 15]
```

```python
plt.plot(
    Day,New_Births,
    linewidth=3,
    linestyle='--',
    marker='o',
    color='pink',
    markerfacecolor='black',
    label='Line 1'
)
plt.title("Birth Rate data of July")
plt.xlabel("day")
plt.ylabel("New_Births")
plt.legend()
plt.show()
```

```python
July_Births = [12.15,11,9,1,9,21, 15]
August_Births = [17,5,2,11.1,8,29, 0]
plt.plot(
    Day,July_Births,
    linewidth=3,
    linestyle='--',
    marker='o',
    color='pink',
    markerfacecolor='black',
    label='July Births'
)
plt.plot(
    Day,August_Births,
    linewidth=3,
    linestyle='--',
    marker='o',
    color='red',
    markerfacecolor='black',
    label='August Births'
)
plt.title("Comparison between the birth rate of July and August")
plt.xlabel("day")
plt.ylabel("New_Births")
plt.legend()
plt.show()
```
